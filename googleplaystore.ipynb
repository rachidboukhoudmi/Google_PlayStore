```python
import numpy as np

def fix_size(s): 
    try:
        s = s.strip()  # Remove leading and trailing whitespace
        if s.endswith('M'):
            return float(s[:-1]) 
        elif s.endswith('k'):
            return float(s[:-1])
        elif s == '1000+':
            return 1_000
        else:
            return np.nan
    except:
        return np.nan 
```

```python
df['Size'] = df['Size'].apply(fix_size)
```

```python
import numpy as np

def fix_price(p):
    try:
        # Ensure the input is a string
        p = str(p).strip()
        if p.startswith('$'):
            return float(p[1:])
        else:
            return 0.0
    except Exception as e:
        print(f"Exception: {e}, Input: {p}")
        return np.nan

```

```python
fix_price('$2.49')
```

```python
df['Price'] = df['Price'].apply(fix_price)
```
